```python
import numpy as np
```

```python
mu = 0
sigma = 0.1
size = 1000

random_x = np.random.normal(mu, sigma, size)
print(random_x[:100])
```

```python
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(random_x, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),       #with equation
         linewidth=5, color='r')

import scipy.stats as stats

x = np.linspace(mu - 3*sigma, mu + 3*sigma, size)
plt.plot(x, stats.norm.pdf(x, mu, sigma), linewidth=2, color='orange')  #with inbuilt function

plt.title('Normal Distribution')
plt.xlabel('Random Variable x')
plt.ylabel('f(x)')
plt.show()
```
